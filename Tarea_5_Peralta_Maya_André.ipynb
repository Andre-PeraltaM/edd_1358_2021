```
class NodoDoble:
    def __init__( self , dato, anterior = None, siguiente = None):
        self.dato = dato
        self.siguiente = siguiente
        self.anterior = anterior

class DoubleLinkedList:
	def __init__(self):
		self.__head = None
		self.__tail = None
		self.__size = 0

	def get_size(self):
		return self.__size

	def is_empty(self):
		return self.__size == 0
		
	def append(self,value):
		if self.is_empty():
			nuevo = NodoDoble(value)
			self.__head = nuevo
			self.__tail = nuevo
		else:
			nuevo = NodoDoble(value,self.__tail,None)
			self.__tail.siguiente = nuevo #tail.next = nuevo
			self.__tail = nuevo
		self.__size += 1

	def transversal(self): #'''recorrido desde head'''
		curr_node = self.__head
		if curr_node != None:
			while curr_node != None :
				print("<--" ,curr_node.dato, "-->" , end="")
				curr_node = curr_node.siguiente
			print('\n')
		else:
			print('Lista vacia')

	def remove_from_head(self,value):
		if self.is_empty():
			pass
		elif self.__head.dato == value:#Inicio
				self.__head = self.__head.siguiente
				self.__head.siguiente.anterior.anterior = None

		else:
			curr_node = self.__head
			while curr_node != None and curr_node.dato != value :
				curr_node = curr_node.siguiente
			if curr_node != None and curr_node.dato == value:
				if curr_node.siguiente == None:#Fin
					curr_node.anterior.siguiente = None
					self.__tail = curr_node.anterior

				else:
					curr_node.anterior.siguiente = curr_node.siguiente
					curr_node.siguiente.anterior = curr_node.anterior
		self.__size -= 1


	def remove_from_tail(self, value ):
		if self.is_empty(): 
			pass
		elif self.__tail.dato == value:#Final
			self.__tail.anterior.siguiente = None
			self.__tail  = self.__tail.anterior

		else:
			curr_node = self.__tail
			while  curr_node != None and curr_node.dato != value :
				curr_node = curr_node.anterior
			if curr_node != None and curr_node.dato == value:
				if curr_node.anterior == None:#Inicio
					curr_node.siguiente.anterior = None
					self.__head = curr_node.siguiente

				else:
					curr_node.anterior.siguiente = curr_node.siguiente
					curr_node.siguiente.anterior = curr_node.anterior
		self.__size -= 1


	def reverse_transversal(self):
		curr_node = self.__tail
		if curr_node != None:
			while( curr_node != None ):
				print("<--" ,curr_node.dato, "-->" , end="")
				curr_node = curr_node.anterior
			print('\n')
		else:
			print('Lista vacia')

ldl = DoubleLinkedList()


ldl.append(60)
ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.append(40)
ldl.append(50)
ldl.append(60)

ldl.remove_from_head(10)
ldl.remove_from_head(40)

ldl.transversal()
ldl.reverse_transversal()

ldl.remove_from_tail(50)
ldl.remove_from_tail(60)


ldl.transversal()
ldl.reverse_transversal()
```
