```
'''
Autor:  Peralta Maya André
Fecha:  04/11/2020
Proposito: 

1. El archivo junio.dat contiene los datos de los trabajadores de una empresa. 

2. La estructura es: número de trabajador, nombres, paterno, materno, horas extra, sueldo base, año de ingreso.

3.Crear un ADT a partir del ADT Array que almacene la información de dicho archivo.

4.Calcular el sueldo para cada uno de ellos tomando en cuenta las siguientes reglas de negocio:
      -La hora extra es pagada a $276.5.
      -Por contrato colectivo de trabajo cada trabajador tiene una prestación del 3% por cada año de antigüedad.

5.Desplegar en pantalla el trabajador con mayor antigüedad y el de menor antiguedad.
'''
class Array:
    def __init__( self , tam ):
        self.datos = [ 0 for x in range(tam)]
    
    def get_item( self ,index ):
        dato = 0
        if index >= 0 and index < len(self.datos):
            dato = self.datos[index]
        else:
            print("Error en el indice") 
        return dato

    def set_item( self  , dato , index  ):
        if index >= 0 and index < len(self.datos):
            self.datos[index] = dato
        else:
            print("Error en el indice")

    def get_length(self):
	    return len(self.datos)
    
    def clear( self , dato ):
        self.datos = [dato for x in range(self.__tam)]

    def __iter__( self ):
    	for i in self.datos:
    		yield i


class Empleado:
    def __init__( self , numero , nombres , paterno , materno , horas_extra ,sueldo_base ,año ):
 
    	try:
	    	if len(numero) ==  4 and numero.isdigit(): #casting de algunos campos
	    		self.__num_trabajador = int(numero)
	    	else:
	    		self.__num_trabajador = None
    	except Exception as e:
    		self.__num_trabajador = None 

    	try:
    		if nombres.isalpha() and not ' ' in nombres: #Nombre
	    		self.__nombres = nombres
	    	else:
	    		self.__nombres = None
    	except Exception as e:
    		self.__nombres = None
    	
    	try:
    		if paterno.isalpha() and not ' ' in paterno:#apellido
	    		self.__paterno = paterno
	    	else:
	    		self.__paterno = None
    	except Exception as e:
    		self.__paterno = None
    	
    	try:
    		if materno.isalpha() and not ' ' in  materno:
	    		self.__materno = materno
	    	else:
	    		self.__materno = None	    		
    	except Exception as e:
    		self.__materno = None
    	
    	try:
    		if int(horas_extra) >= 0 : #horas extras
	    		self.__horas_extra = int(horas_extra)
	    	else:
	    		self.__horas_extra = None
    	except Exception as e:
    		self.__horas_extra =  None
    	
    	try:
    		if int(sueldo_base) >= 0 : #sueldo base
	    		self.__sueldo_base = int(sueldo_base)
	    	else:
	    		self.__sueldo_base = None
    	except Exception as e:
    		self.__sueldo_base = None
    	
    	try:
    		if int(año) <= 2020 and int(año) >= 1938 :
	    		self.__anio_ingreso = int(año)  #Año de entrar a la empresa
	    	else:
	    		self.__anio_ingreso = None
    	except Exception as e:
    		self.__anio_ingreso = None


    def get_num_trabajador(self):
    	return self.__num_trabajador
    def get_nombres(self):
    	return self.__nombres
    def get_paterno(self): 
    	return self.__paterno 
    def get_materno(self): 
    	return self.__materno 
    def get_horas_extra(self): 
    	return self.__horas_extra 
    def get_sueldo_base(self): 
    	return self.__sueldo_base 
    def get_anio_ingreso(self):
    	return self.__anio_ingreso
    def get_salida_estandar(self):
    	return f'Trabajador : {self.__num_trabajador} : {self.__nombres} {self.__paterno} {self.__materno} '

    	#self.__datos = [self.__num_trabajador , self.__nombres, self.__paterno ,self.__materno, self.__horas_extra ,self.__sueldo_base ,self.__anio_ingreso]
    def sueldo_final(self):

    	return (self.__sueldo_base+(self.__sueldo_base*(2020 - self.__anio_ingreso)*0.03)) + (self.__horas_extra*276.5)

    def to_string(self):
    	return [self.__num_trabajador , self.__nombres, self.__paterno ,self.__materno, self.__horas_extra ,self.__sueldo_base ,self.__anio_ingreso] 	    


class Sueldos: #ADT : encapsula datos de nómina 

    def __init__( self ):
        self.__datos = None
    
    def cargar_datos( self , ruta_archivo ): #lee el archivo y lo carga e un ADT Array

        with open(ruta_archivo, 'r',encoding='latin1') as archivo:
        	lineas = archivo.readlines()[1:]
        self.__datos = Array(len(lineas))
        cont = 0
        for i in lineas: #Separar campos
        	dat =  i.split(',')

        	Emp = Empleado(dat[0],dat[1],dat[2],dat[3],dat[4],dat[5],dat[6]) # crear un objeto Empleado
        	self.__datos.set_item(Emp,cont)
        	cont += 1

    def to_string(self):
    	for i in self.__datos:
    		yield i.to_string()

    def calcular_sueldos(self):    # meétodos de uso gral : calculos
    	   
    	for i in self.__datos:
    		yield f'{i.get_salida_estandar()} Sueldo Final : {i.sueldo_final()}'

    def reportar_menor_ant(self):
    	mayor_antiguedad = self.__datos.get_item(0).get_anio_ingreso()
    	empleado = []
    	for i in self.__datos:
    		if i.get_anio_ingreso() > mayor_antiguedad:
    			mayor_antiguedad = i.get_anio_ingreso()
    			empleado = i 

    	return f'El empleado con la menor antiguedad es : {empleado.get_salida_estandar()}con {2020-empleado.get_anio_ingreso()} años en la empresa'

    	
    def reportar_mayor_ant(self):
    	mayor_antiguedad = self.__datos.get_item(0).get_anio_ingreso()
    	empleado = []
    	for i in self.__datos:
    		if i.get_anio_ingreso() < mayor_antiguedad:
    			mayor_antiguedad = i.get_anio_ingreso()
    			empleado = i 
    	return f'El empleado con la mayor antiguedad es {empleado.get_salida_estandar()}con {2020-empleado.get_anio_ingreso()} años en la empresa'

ejemplo = Sueldos()
ejemplo.cargar_datos('junio.dat')

print('\n------------Mayor antiguedad  ----------\n')
print(ejemplo.reportar_mayor_ant())
print('\n------------Menor antiguedad  ----------\n')
print(ejemplo.reportar_menor_ant())

print('\n------------Sueldos finales por trabajador  ----------\n')

for i in ejemplo.calcular_sueldos():
	print(i)

print('\n------------EXTRA : Array de objetos Empleado:  ----------\n')
for i in ejemplo.to_string():
	print(i)
	
```
